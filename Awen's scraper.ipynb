Define a lock, this help launch processes at different times to help out with bandwidth throttling

```python
lock=multiprocessing.Lock()
```

Define functions for each type of page to be scraped

```python
def scrape_this_LA(url):
    try:
        lock.acquire()
        time.sleep(0.100)
        lock.release()
        options=webdriver.ChromeOptions()
        options.add_argument('headless')
        driver = webdriver.Chrome(chrome_options=options)

        driver.get(url)
        time.sleep(3)

        page=driver.page_source
        soup=BS(page,"html.parser")
        wrapper=soup.find(id="template-wrapper")
        driver.quit()
        return [url.split(".co.uk")[1],wrapper.find("h1").text.split(" Local Authority")[0],"Local Authority",wrapper.find("p", {"class":"htmlbox"}).text.split("Local Authority in ")[1].split(" (GSS")[0],wrapper.find("p", {"class":"htmlbox"}).text.split("GSS code ")[1].split(").")[0]]
        print(wrapper.find("h1").text.split(" Local Authority")[0])
    except:
        driver.quit()
```

```python
for i in output:
    if i:
        NOMIS["URL"].append(i[0])
        NOMIS["Place Name"].append(i[1])
        NOMIS["Area Type"].append(i[2])
        NOMIS["Parent area"].append(i[3])
        NOMIS["Area code"].append(i[4])
    else:
        continue
```
